```python
take an image and apply correlation and convolution to it and show both the original and the filtered images.
```

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate2d, convolve2d

# Upload image
uploaded = files.upload()

# Replace with your uploaded image name
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Display original image
plt.figure(figsize=(5,5))
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# Kernel
kernel = np.array([[1, 1],
                   [1, -1]
                   ])

# Correlation
correlation_output = correlate2d(image, kernel, mode='same')

# Convolution
convolution_output = convolve2d(image, kernel, mode='same')

# Show results
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(correlation_output, cmap='gray')
plt.title("Correlation")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(convolution_output, cmap='gray')
plt.title("Convolution")
plt.axis('off')

plt.show()
```
