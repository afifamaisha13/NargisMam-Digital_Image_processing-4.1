```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
color_img = cv2.imread('Lenna_(test_image).png')
```

```python
color_img_rgb = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)

plt.imshow(color_img_rgb)
plt.title("Color Image")
plt.axis('off')
plt.show()
```

```python
gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
gray_direct = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_direct, cmap='gray')
plt.title("Direct Grayscale")
plt.axis('off')
plt.show()
```

```python
normalized_img = color_img / 255.0
```

```python
plt.imshow(normalized_img)
plt.title("Normalized Image")
plt.axis('off')
plt.show()
```

```python
print("Before:", color_img.min(), color_img.max())
print("After:", normalized_img.min(), normalized_img.max())
```
